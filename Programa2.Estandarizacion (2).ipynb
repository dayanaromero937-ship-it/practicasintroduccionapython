```python
from sklearn import preprocessing
import numpy as np
```

### Carga de datos

```python

X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
```

### limpieza de datos 

```python
# Limpieza de datos: estandarización.
standardizer = preprocessing.StandardScaler()

# Estandarizamos los datos de entrenamiento
X_train_std=standardizer.fit_transform(X_train)
print(X_train_std)
```

#### El resultado impreso es la matriz X_train_2_std. Ahora todos los datos están en una escala comparable, lo cual es fundamental para que algoritmos de Machine Learning (como Redes Neuronales o SVM) funcionen correctamente y no den más importancia a una columna simplemente porque tenga números más grandes.

```python
X_train_2 = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
```

```python
# 1. Definir la matriz X_train_2 (Esto es lo que faltaba o no se ejecutó)
X_train_2 = np.array([[ 1., -1.,  2.],
                      [ 2.,  0.,  0.],
                      [ 0.,  1., -1.]])

# 2. Configurar el escalador
f = preprocessing.StandardScaler()

# 3. Ajustar el escalador con los datos de entrenamiento originales
f = f.fit(X_train) 

# 4. Transformar la nueva matriz
X_train_2_std = f.transform(X_train_2)

# 5. Imprimir el resultado
print(X_train_2_std)
```

#### Comprobar si las matrices estandarizadas son iguales ???

```python
son_iguales = np.array_equal(X_train_std, X_train_2_std)
print(f"¿Son exactamente iguales? {son_iguales}")
```

```python
son_iguales_tolerancia = np.allclose(X_train_std, X_train_2_std, rtol=1e-10, atol=1e-10)
print(f"¿Son iguales con tolerancia? {son_iguales_tolerancia}")
```

```python
diferencia = X_train_std - X_train_2_std
print(f"\nDiferencia máxima absoluta: {np.max(np.abs(diferencia))}")
print(f"Diferencia: \n{diferencia}")
```

#### Deber comprobar si son iguales o no

```python
if son_iguales_tolerancia:
    print("\nCONCLUSIÓN: Las dos matrices estandarizadas son iguales.")
    print("   'fit_transform()' es equivalente a hacer 'fit()' seguido de 'transform()'")
else:
    print("\n CONCLUSIÓN: Las matrices NO son iguales.")
```
