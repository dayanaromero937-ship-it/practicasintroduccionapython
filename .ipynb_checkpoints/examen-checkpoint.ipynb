Escribe un programa que muestre por pantalla la concatenación de un número y una cadena de caracteres. Ejemplo: dado el número 3 y la cadena 'abc', el programa mostrará la cadena '3abc'.

```python
numero=3
cadena="abc"
resultado= str(numero)+cadena
print(resultado)
```
