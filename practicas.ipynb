6. Dado un diccionario cuyos elementos son pares de tipo string y numérico (es decir, las claves
son de tipo ‘str’ y los valores son de tipo ‘int’ o ‘float’), escribe un programa que muestre
por pantalla la clave cuyo valor asociado representa el valor númerico más alto de todo el
diccionario. Por ejemplo, para el diccionario {‘a’: 4.3, ‘b’: 1, ‘c’: 7.8, ‘d’:-5} la respuesta
sería ‘c’, dado que 7.8 es el valor más alto de los números 4.3, 1, 7.8 y-5.

```python
# 1. Definimos el diccionario
diccionario = {'a': 4.3, 'b': 1, 'c': 7.8, 'd': -5}

# 2. Buscamos la clave con el valor máximo
# Le decimos a max() que compare usando los valores (.get)
clave_maxima = max(diccionario, key=diccionario.get)

# 3. Mostramos el resultado
print(clave_maxima) # Salida: 'c'
```

7. Dada la lista a = [2, 4, 6, 8] y la lista b = [7, 11, 15, 22], escribe un programa que itere las
listas a y b y multiplique cada elemento de a que sea mayor que 5 por cada elemento de b que
sea menor que 14. El programa debe mostrar los resultados por pantalla

```python
a = [2, 4, 6, 8]
b = [7, 11, 15, 22]

# Una sola línea para filtrar y multiplicar
resultados = [x * y for x in a if x > 5 for y in b if y < 14]

print(resultados) 
# Salida: [42, 66, 56, 88]
```

13. Escribe una dictionary comprehension que construya un diccionario que incluya los tamaños
de cada palabra en una frase dada. Ejemplo: el resultado para la frase “Soy un ser humano”
será {‘Soy’: 3, ‘un’: 2, ‘ser’: 3, ‘humano’: 6}

```python
frase = "Soy un ser humano"

# 1. Separamos la frase en palabras usando .split()
# 2. Creamos el diccionario: palabra (clave) y len(palabra) (valor)
resultado = {palabra: len(palabra) for palabra in frase.split()}

print(resultado)
# Salida: {'Soy': 3, 'un': 2, 'ser': 3, 'humano': 6}
```

14. Escribe una list comprehension que construya una lista que incluya todos los números del 1
al 10 en orden. La primera mitad se mostrarán en formato numérico; la segunda mitad en
texto. Es decir, el resultado será: [1, 2, 3, 4, 5, ‘seis’, ‘siete’, ‘ocho’, ‘nueve’, ‘diez’]

```python
nombres = ['', 'uno', 'dos', 'tres', 'cuatro', 'cinco', 'seis', 'siete', 'ocho', 'nueve', 'diez']

# Si n <= 5, guarda n. Si no, busca el nombre en la lista 'nombres'.
resultado = [n if n <= 5 else nombres[n] for n in range(1, 11)]

print(resultado)
# Salida: [1, 2, 3, 4, 5, 'seis', 'siete', 'ocho', 'nueve', 'diez']
```
