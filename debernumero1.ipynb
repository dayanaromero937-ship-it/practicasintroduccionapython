2. Desarrollar un programa en python donde se resuelva una ecuacin de segundo
grado mediante la frmula general, el usuario debe ingresar por teclado los valores
a,b y c y mostrar en pantalla los valores de la ecuacin.

```python
import math

# Entrada de datos
a = float(input("Valor de a: "))
b = float(input("Valor de b: "))
c = float(input("Valor de c: "))

# Clculo del discriminante (la parte dentro de la raz)
discriminante = b**2 - 4*a*c

if discriminante < 0:
    print("La ecuacin no tiene soluciones reales.")
else:
    # Clculo de las dos soluciones
    x1 = (-b + math.sqrt(discriminante)) / (2*a)
    x2 = (-b - math.sqrt(discriminante)) / (2*a)

    print(f"Resultados: x1 = {x1}, x2 = {x2}")
```

Desarrollar un programa en python donde el usuario ingrese dos numeros por teclado el primer numero va a ser el inicio de un rango el segundo numero va a ser el final de dicho rango mediante el ciclo while verificar cuantos numeros son pares y mostrar los numeros y el total de numero pares

```python
## #ingresamos en letras los dos numeros por teclado que nos pide del rango
numeros = int(input("ingrese el primer numero del rango: "))
numero2 = int(input("ingrese el segundo numero del rango: "))
contador = 0

while numeros < numero2:
    if numeros % 2 == 0:
        print(numeros)
        contador += 1
    # Esta lnea debe estar alineada con el 'if'
    numeros += 1

print("el total de numeros pares es", contador)
```

Desarrolle un programa en python donde permita al usuario ingresar un numero entero positivo repetitivamente el programa debe continuar mientras se ingrese un numero mayor a cero. Al finalizar el proceso debe mostrarme la suma de los numero ingresados

```python
#total acumulado
suma = 0
#inicia un bucle que se repite sin fin
while True:
    try:#convierte lo que se escribe en un numero
        dato = int(input('Ingrese un numero: '))
        if dato < 0: break
        suma += dato #suma del numero total
    except ValueError:
        break #si se deja vacio se vuelve un bucle infinito

print("La suma es:", suma)
```

Crear un programa en python donde el usuario ingrese un numero por tecalo verificar si un numero es primo usando whyle

```python
n = int(input("Nmero: "))
d = 2 # Divisor inicial
primo = True if n > 1 else False # 1 o menores no son primos

while d < n and primo: # Se ejecuta mientras d sea menor a n y siga siendo primo
    if n % d == 0:
        primo = False # Si el resto es 0, encontramos un divisor
    d += 1 # Aumentamos el divisor para evitar el bucle infinito

print("Es primo" if primo else "No es primo")
```
