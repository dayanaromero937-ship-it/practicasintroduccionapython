Dada una lista con elementos duplicados, escribir un programa que muestre una nueva lista con el mismo contenido que la primera pero sin elementos duplicados. Para este ejercicio, no puedes hacer uso de objetos de tipo ‘Set’

```python
lista_duplicados=[2,2,2,2,3,3,5,6,7,2]
lista_nueva=[]
for numero in lista_duplicados:
    if numero not in lista_nueva:
        lista_nueva.append(numero)
print(lista_nueva)
```

5. Dada una lista L de números positivos, escribir un programa que muestre otra lista (ordenada)
que contenga todo índice i que cumpla la siguiente condición: L[i] es múltiplo de 3. Por
13
ejemplo, dada la lista L = [3,5,13,12,1,9] el programa mostrará la lista [0,3,5] dado que L[0],
L[3] y L[5] son, respectivamente, 3, 12 y 9, que son los únicos múltiplos de 3 que hay en L.

```python
l=[3,5,13,12,1,9,]
resultado = [i for i in range(len(l)) if l[i] % 3 == 0]
print(resultado)  
```

6. Dado un diccionario cuyos elementos son pares de tipo string y numérico (es decir, las claves
son de tipo ‘str’ y los valores son de tipo ‘int’ o ‘float’), escribe un programa que muestre
por pantalla la clave cuyo valor asociado representa el valor númerico más alto de todo el
diccionario. Por ejemplo, para el diccionario {‘a’: 4.3, ‘b’: 1, ‘c’: 7.8, ‘d’:-5} la respuesta
sería ‘c’, dado que 7.8 es el valor más alto de los números 4.3, 1, 7.8 y-5

```python
diccionario={"a": 4.3, "b": 1, "c": 7.8, "d":-5}
clave_maxima=max(diccionario,key=diccionario.get)
print(clave_maxima)
```

7. Dada la lista a = [2, 4, 6, 8] y la lista b = [7, 11, 15, 22], escribe un programa que itere las
listas a y b y multiplique cada elemento de a que sea mayor que 5 por cada elemento de b que
sea menor que 14. El programa debe mostrar los resultados por pantalla.

```python
a= [2, 4, 6, 8]
b = [7, 11, 15, 22]
resultado=[x*y for x in a if x >5 for y in b if y <14]
print(resultado)
```

Escribe una list comprehension que construya una lista con los números enteros positivos de una lista de números dada. La lista original puede incluir números de tipo float, los cuales deben ser descartados.

```python
lista_mezclada = [3, -5, 4.5, 10, -2.1, 7, 0]
resultado = [x for x in lista_mezclada if type(x) == int and x > 0]

print(resultado) 
```

Escribe una dictionary comprehension que construya un diccionario que incluya los tamaños de cada palabra en una frase dada. Ejemplo: el resultado para la frase “Soy un ser humano” será {‘Soy’: 3, ‘un’: 2, ‘ser’: 3, ‘humano’: 6}

```python
frase = "Soy un ser humano"
resultado = {palabra: len(palabra) for palabra in frase.split()}
print(resultado)
```
