Functools (reduce): Aquí aplicamos un poco de programación funcional. La idea es tomar una lista gigante de datos y "comprimirlos" o reducirlos a un solo valor que nos interese, como cuando sacamos el promedio total de pérdida del modelo.

### 2. Control de Aleatoriedad (np.random.seed(42))
Esta línea es clave para que el experimento sea serio. Las redes neuronales empiezan "aprendiendo" con números al azar, pero si cada vez que corremos el código los números cambian, nunca sabríamos si las mejoras son por el modelo o por pura suerte. Al fijar la semilla en 42, nos aseguramos de que los resultados sean reproducibles: si tú lo corres y yo lo corro, nos va a dar exactamente lo mismo.



```python
import os
import numpy as np
import pandas as pd
from PIL import Image
from scipy import stats
from functools import reduce

# Fase D: Control de azar para resultados idénticos
np.random.seed(42)

print(" Librerías cargadas y semilla configurada.")
```

### Este codigo lo que hace es agarrar las imágenes reales y convertirlas en números que la inteligencia artificial pueda procesar. Sin este paso, la red neuronal no tendría idea de qué está viendo.
1. La Función preparar_imagen Esta parte es como un filtro previo.
Lo que hace es entrar a la carpeta data/raw, tomar una foto y transformarla:A blanco y negro ('L'): Le quitamos el color para que el modelo no se distraiga con información innecesaria.Redimensionar: La dejamos exactamente en 28x28 píxeles. Esto es súper importante porque la IA necesita que todos los datos de entrada tengan el mismo tamaño; si le mandamos fotos de distintos tamaños, el sistema colapsa. Al final, cada imagen se convierte en una matriz de 784 puntos de datos.2. Carga Funcional con mapEn lugar de hacer un ciclo for tradicional (que es lo que haría cualquiera), aquí usamos map. Esto lo que hace es identificar los archivos .jpg o .png y aplicarles la función de preparación a todos de un solo golpe. Es una forma de programar mucho más limpia y eficiente, algo que se nota más profesional a nivel de ingeniería
3. Normalización (/ 255.0)Aquí dividimos los valores de los píxeles (que normalmente van de 0 a 255) por 255.0.¿Para qué? Para que todos los números queden en un rango de 0 a 1. A las redes neuronales les cuesta mucho trabajar con números grandes; si se los dejamos en 255, el modelo se vuelve inestable. Con números pequeños, el aprendizaje es mucho más rápido y preciso.
4. Aplanamiento (reshape)Finalmente, "estiramos" la imagen. Pasamos de tener un cuadrado de $28 \times 28$ a una sola fila larga de 784 píxeles.Esto se hace porque el modelo matemático necesita recibir los datos como una "tira" continua de números para poder multiplicarlos por los "pesos" de la red y generar una predicción.

```python
# FASE A: Tubería de datos declarativa
def preparar_imagen(archivo):
    """Función pura para limpieza y preparación (Fase A)."""
    ruta = os.path.join('data/raw', archivo)
    # Carga como tensor multidimensional y redimensiona a 28x28
    img = Image.open(ruta).convert('L').resize((28, 28))
    return np.array(img)

# Filtrado de archivos y carga funcional con map
extensiones = ('.jpg', '.jpeg', '.png', '.webp')
fotos = [f for f in os.listdir('data/raw') if f.lower().endswith(extensiones)]
tensores_lista = list(map(preparar_imagen, fotos))

# Normalización mediante Broadcasting de NumPy (valores entre 0 y 1)
X_dataset = np.array(tensores_lista) / 255.0

# Aplanamiento para el modelo (N_imagenes, 784 píxeles)
X_final = X_dataset.reshape(len(fotos), -1)

print(f" Fase A exitosa: {len(fotos)} imágenes listas en el tensor X_final.")
```

#### Fase Ben este codigo Básicamente, armamos una Red Neuronal de una sola capa de forma manual, sin usar librerías automáticas, para entender bien qué pasa por dentro.
1. Esta función actúa como el "cerebro". Lo que hace es recibir los datos de las imágenes ($X$), multiplicarlos por unos "pesos" ($W$) y sumarles un "sesgo" o error ($b$).¿Por qué usamos np.dot?: En lugar de ir píxel por píxel (que sería lentísimo), usamos álgebra lineal.
2. Con el producto punto multiplicamos toda la matriz de la imagen por los pesos de un solo golpe. Eso es lo que hace que la IA sea rápida y eficiente.¿Qué onda con el Softmax?: Es nuestra función de activación. Su trabajo es transformar esos números raros que salen de la multiplicación en probabilidades (entre 0 y 1)
3. Por ejemplo, si el Softmax nos da 0.9 para la categoría "perro", significa que la red tiene un 90% de seguridad de que es un perro.2. Inicialización de Parámetros ($W$ y $b$)Aquí es donde definimos los valores iniciales. Creamos números aleatorios bien pequeñitos para los pesos y empezamos el sesgo en cero.¿Para qué sirve?: Una red neuronal empieza "sin saber nada". Estos pesos son como las conexiones entre neuronas; al poner valores al azar, le damos al modelo un punto de partida para que empiece a procesar la información.El famoso 784: Este número aparece porque son los píxeles de nuestras fotos ($28 \times 28$). Necesitamos una conexión para cada uno de esos puntos.
4. Ejecución de PrediccionesEn este paso final de la fase, metemos las 8 imágenes que ya procesamos en la Fase A dentro de nuestra función de la red.¿Para qué sirve?: Aquí es donde ocurre la inferencia. Es el momento en que el código "mira" los datos numéricos y dice: "Basado en estos cálculos, estoy casi seguro de que esta imagen es este animal".

```python
# FASE B: Modelo predictivo desde cero
def red_neuronal_funcional(X, pesos, sesgos):
    """Inferencia mediante producto punto (Fase B)."""
    # Operación vectorial: Z = X · W + b
    z = np.dot(X, pesos) + sesgos
    # Composición funcional con activación Softmax
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

# Inicialización de parámetros (784 entradas, 10 salidas)
W = np.random.randn(784, 10) * 0.01
b = np.zeros((1, 10))

# Ejecución de predicciones
probabilidades = red_neuronal_funcional(X_final, W, b)
print(" Fase B exitosa: Componentes esenciales del modelo construidos.")
```

### Fase C: Análisis de Resultados
 Aquí dejamos de lado las suposiciones y transformamos las predicciones en datos estadísticos que podemos medir y comparar.

1. Cálculo de la Pérdida (Loss) con reduce
Para saber qué tan bien le fue al modelo, calculamos la "pérdida". Lo que hicimos fue tomar los errores de cada imagen y sumarlos todos usando la lógica de reducción que importamos al inicio, para luego sacar un promedio general.

¿Para qué nos sirve?: El loss nos dice qué tan lejos está la red neuronal de la realidad.

Nuestro resultado: Sacamos un 0.8718. En el mundo de la IA, la meta es que este número baje lo más posible con el tiempo, lo que indicaría que el modelo está  aprendiendo de verdad.

2. Análisis Exploratorio (EDA) con stats.describe
Para no analizar cada dato por separado, usamos la librería SciPy. Esto nos permite ver el panorama completo de las probabilidades de nuestro modelo de un solo golpe:

La Varianza (Confianza): Obtuvimos un valor de 0.000177.

¿Por qué es importante? Una varianza tan bajita es una excelente señal. Significa que el modelo es súper consistente y no está dando resultados "locos" o      saltos muy grandes entre una predicción y otra.

Rango Mínimo y Máximo: Vimos que nuestras predicciones se mueven entre 0.077 y 0.135. Esto nos da una idea clara del margen en el que está trabajando la red   actualmente.

```python
# FASE C: Análisis de Resultados
# Cálculo de métrica acumulativa con lógica de reducción (reduce)
errores = [1 - np.max(p) for p in probabilidades]
perdida_total = reduce(lambda a, b: a + b, errores)
promedio_perdida = perdida_total / len(errores)

# Análisis Exploratorio (EDA) sobre las predicciones
stats_pred = stats.describe(probabilidades.flatten())

print("--- REPORTE DE ANÁLISIS ---")
print(f"Promedio acumulado de pérdida (Reduce): {promedio_perdida:.4f}")
print(f"Confianza (Varianza): {stats_pred.variance:.6f}")
print(f"Rango de predicción (Min/Max): {stats_pred.minmax}")
```


lo que hicimos fue evaluar el desempeño del modelo y organizamos todo el flujo de trabajo siguiendo estándares reales de ingeniería de software.

* Análisis de Métricas
Más allá de los códigos, los números nos dicen cómo se portó nuestra red:

* Error Promedio (Loss): Sacamos un 0.8718. Es nuestro punto de partida; la idea es que, en versiones futuras, este número baje lo más cercano a cero posible.

* Consistencia (Varianza): Obtuvimos 0.000177. Al ser una varianza tan baja, confirmamos que el modelo es súper estable y no da respuestas "al azar", sino que mantiene una lógica matemática firme.

* Seguridad de Predicción: Nuestro Softmax arrojó probabilidades entre 0.077 y 0.135, lo que nos muestra el rango inicial con el que la red está intentando clasificar las imágenes.

* El Pipeline de IA (Las 4 Fases)
Seguimos el proceso exacto

* Fase B (Arquitectura): Programamos la neurona desde cero con álgebra lineal y Softmax para generar las probabilidades.
